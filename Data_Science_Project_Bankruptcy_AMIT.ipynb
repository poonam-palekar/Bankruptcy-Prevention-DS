```python
### EDA: Overall distributions (all companies)
#Show overall counts at 0/0.5/1 for each risk feature to understand baseline distribution across all companies
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
features = ['industrial_risk','management_risk','financial_flexibility',
            'credibility','competitiveness','operating_risk']

sns.set_theme(style="whitegrid", font_scale=1.05)
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
for ax, col in zip(axes.flat, features):
    sns.countplot(x=col, data=df, palette='coolwarm', ax=ax)
    ax.set_title(col.replace('_',' ').title())
    ax.set_xlabel('Value (0=Low, 0.5=Medium, 1=High)')
    ax.set_ylabel('Count')
plt.suptitle('Feature Distributions', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
## EDA: Class-wise distributions (bankruptcy vs non-bankruptcy)
#Compare counts at 0/0.5/1 by class to see where bankruptcies concentrate vs non‑bankruptcies for each feature
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
for ax, col in zip(axes.flat, features):
    sns.countplot(x=col, hue='class', data=df, palette='coolwarm', ax=ax)
    ax.set_title(col.replace('_',' ').title())
    ax.set_xlabel('Value (0, 0.5, 1)')
    ax.set_ylabel('Count')
    ax.legend(title='Class', loc='upper right')
plt.suptitle('Feature Distributions by Class', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
##Build a compact table summarizing counts and bankruptcy rates at 0/0.5/1 for each feature to capture key EDA insights
```

```python
summary = []
for col in features:
    dist = df[col].value_counts().sort_index().to_dict()
    rate = df.groupby(col)['class'].apply(lambda x: (x=='bankruptcy').mean()).round(3).to_dict()
    summary.append({'feature': col, 'counts': dist, 'bankruptcy_rate': rate})
pd.DataFrame(summary)
```
