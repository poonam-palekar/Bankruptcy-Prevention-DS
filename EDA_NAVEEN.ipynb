```python
## separating the features and the target
feature=df.drop(columns=['class'])
feature
```

```python
target=df[['class']]
target
```

```python
#Finding the correlation between the features
corr=feature.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#Checking for multicolinearity using variance_inflation_factor
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
VIF=pd.DataFrame()
VIF['Features']=feature.columns
VIF
```

```python
VIF['vif']=[variance_inflation_factor(feature.values,i) for i in range(len(feature.columns))]
```

```python
VIF.sort_values(by='vif',ascending=False)
```

```python
#Labeling the target
label=LabelEncoder()
target['class']=label.fit_transform(target['class'])
target
```

```python
#Checking the correlation between the features and the target
df=feature.join(target)
df
```

```python
corr_ft=df.corr()
sns.heatmap(corr_ft,annot=True)
plt.show()
```

```python
#Finding the relation using annova
from sklearn.feature_selection import f_classif
```

```python
fcls=f_classif(feature,target)
pd.Series(fcls[0],index=feature.columns).sort_values(ascending=False).plot(kind='bar')
plt.title('Realtion between Feature and Target')
plt.xlabel('Features')
plt.show()
```
