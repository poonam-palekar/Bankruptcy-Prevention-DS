```python
#Labeling the target
label=LabelEncoder()
target['class']=label.fit_transform(target['class'])
target
```

```python
corr_ft=df.corr()
sns.heatmap(corr_ft,annot=True)
plt.show()
```

```python
#Finding the relation using annova
from sklearn.feature_selection import f_classif
```

```python
fcls=f_classif(feature,target)
pd.Series(fcls[0],index=feature.columns).sort_values(ascending=False).plot(kind='bar')
plt.title('Realtion between Feature and Target')
plt.xlabel('Features')
plt.show()
```

```python
## Splitting the dataset into training data and testing data
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,target,train_size=0.75,random_state=100)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Logistic Regression Model

```python
#importing the Logistic Regression model and the metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
log=LogisticRegression()
```

```python
#Training the model
log.fit(x_train,y_train)
```

```python
#Predicting the target for testing features
y_pred=log.predict(x_test)
```

```python
# Comparing the predicted and the actual value
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred))
print("Recall:",recall_score(y_test,y_pred))
print("F1_Score:",f1_score(y_test,y_pred))
```

```python
# Getting the predicted probability values
sigmoid=log.predict_proba(x_test)[:,1]
sigmoid
```

## ROC Curve

```python
# Importing the ROC Curve
from sklearn.metrics import roc_auc_score,roc_curve
auc_score=roc_auc_score(y_test,sigmoid)
```

```python
auc_score
```

```python
fpr,tpr,threshold=roc_curve(y_test,sigmoid)
```

```python
plt.plot(fpr,tpr,lw=1.5,label=f"AUC_SCORE:{auc_score}")
plt.plot([0,1])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC CURVE")
plt.grid()
plt.legend()
plt.show()
```

## KNN Model
